## IMPORTAMOS LOS PAQUETES

```python
import requests as rq 
import os
import pandas as pd
```

## FUNCIÓN QUE EXTRAE LOS DF

```python
tipo = "mh"
for file in os.listdir("./Municipios"):
    print(file)
    for year in [i for i in range(2023, 2016,-1)]:
        print(year)
        try:
            nfile = file[:-8] + str(year) + ".txt"
            print(nfile)
            petition = rq.get("https://webcat-web.gva.es/webcat_web/datosHistoricosRvvcca/descargarHistorico?nomFichero="+nfile+"&red=&anyo="+str(year)+"&tipoFich=mh")
            print(petition.status_code)
            if petition.status_code == 200:
                open(f'./Datos/{nfile}', 'wb').write(petition.content)
            else:
                break
        except:
            break
```

## UNIFICAMOS DF 

```python
path='C:/Users/annal/Desktop/TFG/Datos'
dirs=os.listdir(path)
dirs=list(dirs)
# dirs
```

```python
with open(path+'/'+dirs[0]) as f:
    est = f.readlines()[1]
#print(est)
    estacion=str.strip(est[19:])
#print(estacion)
```

```python
df = pd.read_csv(path+'/'+dirs[0], encoding="ISO 8859-15" ,sep="\t",skiprows=(0,1,2,4))
df['Estación']=estacion
df
```

```python
for file in range(1,len(dirs)):
    
    #print(path+'/'+dirs[file])
    try:
        with open(path+'/'+dirs[file]) as f:
            est = f.readlines()[1]
            estacion=str.strip(est[19:])
            #print(estacion)
            df2 = pd.read_csv(path+'/'+dirs[file], encoding="ISO 8859-15" ,sep="\t",skiprows=(0,1,2,4))
            df2['Estación']=estacion
            df=pd.concat([df, df2],sort=False)
    except: 
        pass
df
```

```python
# comprobamos el número de faltantes 

for i in df.columns:
    print(f'La columna {i} tiene {df[i].isna().sum()} valores faltantes')
```

```python
## creamos el df

df.to_csv('data2017_2022.csv',decimal='.',header=True, index=False)
```
